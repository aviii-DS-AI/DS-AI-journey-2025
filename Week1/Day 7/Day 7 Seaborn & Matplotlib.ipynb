âœ… Why we visualize data â€” when tables arenâ€™t enough

Tables show raw data; visualizations help you:

-Spot trends and patterns quickly.
-Compare values easily.
-Identify outliers or anomalies.
-Tell a story or communicate findings effectively.

 Example: A dataset of monthly sales in a table vs. a line graph â€” the graph instantly shows seasonality or growth.

âœ… Intro to matplotlib.pyplot


Matplotlib, which is a widely used library for creating static, animated, and interactive visualizations in Python.
It's often used for data analysis, scientific computing, and research.



```python
import matplotlib.pyplot as plt

# plot() â€“ Line Plot
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()

# scatter() â€“ Scatter Plot
a = [1, 2, 3, 4]
b = [10, 20, 25, 30]

plt.scatter(a, b)
plt.title("Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()

# bar() â€“ Bar Chart
categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values)  # Create a vertical bar chart
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()


# hist() â€“ Histogram
data = [10, 20, 20, 30, 40, 40, 40, 50, 60]

plt.hist(data, bins=6) # Create histogram with 6 bins/bars
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()

```

2. Introduction to seaborn
Seaborn, a Python library for data visualization. Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics.




```python

import seaborn as sns
import matplotlib.pyplot as plt           #ðŸ§ª Basic Setup
import pandas as pd
df = sns.load_dataset("titanic")  # Loads data set from online seaborn library

#  sns.countplot() â€“ Count of Categorical Data

sns.countplot(x="class", data=df) # Count the number of entries per class
plt.title("Passenger Class Count")
plt.show()

# sns.boxplot() â€“ Box Plot for Distributions
sns.boxplot(x="class", y="age", data=df)  # Show age distribution per class
plt.title("Age Distribution by Class")
plt.show()

# sns.heatmap() â€“ Correlation Heatmap  A heatmap is a color-coded matrix that shows the strength
                                 #     of relationships (usually correlation) between numerical features.
corr = df.corr(numeric_only=True)  # Calculate numeric correlations

sns.heatmap(corr, annot=True, cmap="coolwarm")  # Create heatmap with annotations
plt.title("Correlation Heatmap")
plt.show()

# ------- Pair Plot --------A grid of scatter plots showing relationships between all pairs of features in a dataset. 
#It can also include histograms or KDEs on the diagonal.

iris = sns.load_dataset("iris")  # Load |iris| dataset

sns.pairplot(iris, hue="species")  # Plot pairwise relationships, colored by species
plt.suptitle("Pairplot of Iris Dataset", y=1.02)  # Title above all plots
plt.show()






```

```python
#   Practice Tasks
# Step 1: Load the Titanic Dataset------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
df = sns.load_dataset('titanic')

sns.set_style("whitegrid")  # Clean background for all plots


# Optional: Check first few rows
print(df.head(5))

# Step 2: Bar Chart of Survival Count---------------------------------------

# Bar chart: Count of Dead vs Survived
sns.countplot(x='survived', data=df)

# Add titles and labels
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()

# Step 3: Histogram of Passenger Ages---------------------------------------

# Histogram of ages
plt.hist(df['age'].dropna(), bins=20, color='skyblue', edgecolor='black')

plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Step 4: Scatter Plot (Age vs Fare)
# Scatter plot: Age vs Fare
plt.scatter(df['age'], df['fare'], alpha=0.6, color='green')

plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()

# Step 5: Heatmap of Correlations
# Compute correlation matrix (numerical only)
corr = df.corr(numeric_only=True)

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()















```

Analysis Goals
Goal 1: Age & Survival



```python
# Goal 1: Age & Survival
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="survived", y="age", data=df)

plt.title("Age vs Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()

#ðŸ“ Insight:
#Younger passengers (especially children) had a higher chance of survival.
#The median age of survivors appears slightly lower compared to non-survivors.


# Goal 2: Fare & Survival

sns.boxplot(x="survived", y="fare", data=df)

plt.title("Fare vs Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.yscale("log")  # Fares have outliers; log scale makes it readable
plt.show()

#ðŸ“ Insight:
#Passengers who paid higher fares were more likely to survive. This suggests that class or cabin location may have influenced survival chances.


# Goal 3: Gender & Survival
sns.countplot(x="sex", hue="survived", data=df)

plt.title("Gender vs Survival")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
#ðŸ“ Insight:
#A much higher proportion of women survived compared to men. This supports the â€œwomen and children firstâ€ policy during evacuation.




```
